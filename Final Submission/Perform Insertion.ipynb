```python
def convert_to_int_or_null(x):
    if pd.notna(x):
        return int(x)
    else:
        return 'NULL'
```

```python
def convert_to_int(x):
    if pd.notna(x):
        return int(x)
    elif x == NULL:
        pass
```

```python
df['ConferenceID'] = df['ConferenceID'].apply(convert_to_int_or_null)
```

```python
df['OpponentID'] = df['OpponentID'].apply(convert_to_int_or_null)
```

```python
df['LocationID'] = df['LocationID'].apply(convert_to_int_or_null)
```

```python
tuple_rows_with_null = df.apply(lambda row: (row['OpponentID'], row['LocationID'], row['ConferenceID']), axis=1)
```
